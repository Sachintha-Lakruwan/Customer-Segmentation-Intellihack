```python
import numpy as np
import pandas as pd

customers = pd.read_csv("customer_behavior_analytcis.csv", index_col="customer_id")
customers.head()
```

```python
features = ['total_purchases', 'avg_cart_value', 'total_time_spent', 'product_click', 'discount_counts']
customers.isnull().sum()
```

```python
customers.shape
```

```python
customers = customers.dropna(subset = features)
```

```python
data = customers[features].copy()
```

```python
data.head()
```

```python
data = (data - data.min()) / (data.max() - data.min()) * 9 + 1
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(3)
kmeans.fit(data)
```

```python
centroids = kmeans.cluster_centers_
pd.DataFrame(centroids, columns=features).T
```

1. Cluster 0 => Bargain Hunters
2. Cluster 1 => Window Shoppers
3. Cluster 2 => High Spenders

```python
data["cluster"] = kmeans.labels_
```

```python
data
```

**Bargain Hunters**

```python
bargain_hunters = customers[data.cluster == 0]
bargain_hunters
```

**Window Shoppers**

```python
window_shoppers = customers[data.cluster == 1]
window_shoppers
```

**High Spenders**

```python
high_spenders = customers[data.cluster == 2]
high_spenders
```

```python
submission = data[["cluster"]]
submission
```

```python
submission['cluster'] = submission['cluster'].replace({
    0: 'Bargain Hunter',
    1: 'Window Shopper',
    2: 'High Spender'
})
submission
```

```python
submission.to_csv('predictions.csv')
```
